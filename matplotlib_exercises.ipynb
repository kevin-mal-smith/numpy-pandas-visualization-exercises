### Exercise 1:
Use matplotlib to plot the following equation:

### Exercise 4:
Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
plt.figure(figsize = (10,10))


x = list(range(1,5))
y1 = [n**.5 for n in x]
y2 = [n**3 for n in x]
y3 = [2**n for n in x]
y4 = [(1/(n+1)) for n in x]

plt.plot(x,y1,c = 'blue',label ='$√x$')
plt.plot(x,y2, c = 'green',label ='$x^3$')
plt.plot(x,y3, c = 'red',label ='$2^x$' )
plt.plot(x,y4, c='m',label='$1/(x+1)$')

plt.ylim(0,15)
plt.legend()
plt.title('Various Equations')
```
