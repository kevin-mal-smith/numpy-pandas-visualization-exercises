```python
import numpy as np
import matplotlib.pyplot as plt
import math

plt.figure(figsize=(10,10))
x = list(range(1,50))
y1 = [1 for n in x]
y2 = [math.log(n) for n in x]
y3 = [n for n in x]
y4 = [n * math.log(n) for n in x]
y5 = [n**2 for n in x]
y6 = [2**n for n in x]
y7 = [math.factorial(n) for n in x]
y8 = [n**n for n in x]


plt.plot(x,y1, label = '$O(1)$')
plt.plot(x,y2, label = '$O(\logn)$')
plt.plot(x,y3, label = '$O(n)$')
plt.plot(x,y4, label = '$O(n \logn)$')
plt.plot(x,y5, label = '$O(n^2)$')
plt.plot(x,y6, label = '$O(2^n)$')
plt.plot(x,y7, label = '$O(n!)$')
plt.plot(x,y8, label = '$O(n^n)$')
plt.legend()

plt.xlim(1,20)
plt.ylim(0,30)
plt.title('Big O Nation')
plt.xlabel('Element')
plt.ylabel('Operations')
```
